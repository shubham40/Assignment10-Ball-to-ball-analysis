```python
#1. Find the top 5 batsman who have hit the most number of 6's in the last 5 overs in the entire IPL history
```

```python
import numpy as np
import pandas as pd
```

```python
delivery=pd.read_csv('deliveries.csv')
```

```python
delivery.head()
```

```python
mask2=delivery[delivery['over']>15]  ## death over
```

```python
mask3=mask2[mask2['batsman_runs']==6]
```

```python
mask3.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head()
```
