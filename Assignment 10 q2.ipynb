```python
import numpy as np
import pandas as pd
```

```python
delivery=pd.read_csv('deliveries.csv')
```

```python
delivery.head()
```

```python
mask4=delivery[delivery['batsman']=='V Kohli']
mask4
```

```python
mask4.groupby('bowler')['batsman_runs'].sum().sort_values(ascending=False).head()
```
